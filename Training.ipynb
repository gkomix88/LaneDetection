```
import os
import glob
from pathlib import Path

os.mkdir('train')
os.mkdir('valid')
os.mkdir('test')

os.mkdir('train/images')
os.mkdir('valid/images')
os.mkdir('test/images')

os.mkdir('train/labels')
os.mkdir('valid/labels')
os.mkdir('test/labels')

num_set = len(os.listdir('export/images/'))
num_lab = len(os.listdir('export/labels/'))

train_name = os.listdir('export/images/')[:int(0.7*num_set)]
valid_name = os.listdir('export/images/')[len(train_name):len(train_name) + int(0.2*num_set)]
test_name = os.listdir('export/images/')[len(train_name) + len(valid_name):]

train_lab = os.listdir('export/labels/')[:int(0.7*num_lab)]
valid_lab = os.listdir('export/labels/')[len(train_lab):len(train_lab) + int(0.2*num_lab)]
test_lab = os.listdir('export/labels/')[len(train_lab) + len(valid_lab):]

for i in range(num_set):
    if i < len(train_name):
        Path("export/images/"+train_name[i]).rename("train/images/"+train_name[i])
    if i < len(valid_name):
        Path("export/images/"+valid_name[i]).rename("valid/images/"+valid_name[i])
    if i < len(test_name):
        Path("export/images/"+test_name[i]).rename("test/images/"+test_name[i])
        
for i in range(num_lab):
    if i < len(train_lab):
        Path("export/labels/"+train_lab[i]).rename("train/labels/"+train_lab[i])
    if i < len(valid_lab):
        Path("export/labels/"+valid_lab[i]).rename("valid/labels/"+valid_lab[i])
    if i < len(test_lab):
        Path("export/labels/"+test_lab[i]).rename("test/labels/"+test_lab[i])
```
